```python
mcu_steve['compound'] = ''
mcu_steve['negative'] = ''
mcu_steve['neutral'] = ''
mcu_steve['positive'] = ''
mcu_steve.head()
```

```python
 import nltk
 nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import unicodedata

# instantiating the Sentiment Analyzer
sid = SentimentIntensityAnalyzer()

# calculating sentiment scores
mcu_steve['compound'] = mcu_steve['line'].apply(lambda x: sid.polarity_scores(x)['compound'])
mcu_steve['negative'] = mcu_steve['line'].apply(lambda x: sid.polarity_scores(x)['neg'])
mcu_steve['neutral'] = mcu_steve['line'].apply(lambda x: sid.polarity_scores(x)['neu'])
mcu_steve['positive'] = mcu_steve['line'].apply(lambda x: sid.polarity_scores(x)['pos'])

# displaying the stock data
mcu_steve.head()
```

```python
mcu_steve['sentiment']=0;
```

```python
mcu_steve.head()
```

```python
mcu_steve=mcu_steve.reset_index()
```

```python
mcu_steve.drop(['index'],axis=1,inplace=True)
```

```python
for row in range(0,len(mcu_steve)):
  if mcu_steve['compound'][row]< -0.5:
    mcu_steve['sentiment'][row]=-1
  elif mcu_steve['compound'][row]>0.5:
    mcu_steve['sentiment'][row]=1
  else:
    mcu_steve['sentiment'][row]=0
mcu_steve.head()
```
